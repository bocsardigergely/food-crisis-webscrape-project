```python
import requests
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
import time
```

```python
url = "../spar-website.html"
coverpage = open(url, encoding="utf8")
soup1 = BeautifulSoup(coverpage.read(), features="lxml")
coverpage_news = soup1.find_all('div', class_="col-sm-6 col-md-4")
```

```python
links = []
dates = []
titles = []
for l in range(len(coverpage_news)):
    links.append(coverpage_news[l].find('a')['href'])
    dates.append(coverpage_news[l].find('p').get_text())
    titles.append(coverpage_news[l].find('h3').get_text())

len(links)
```

```python
cikkek_adatai = []
start = time.time()
counter = 0
for m in links:
    # link
    article = requests.get(m)
    article_content = article.content
    soup_article = BeautifulSoup(article_content, 'html5lib')

    # cím
    title = titles[counter]

    #date
    date = dates[counter]

    # törzsszöveg
    body = soup_article.find_all('div', class_="col-sm-8 col-sm-offset-2")
    body = body[2]
    final_article = body.get_text().replace("\n", " ")
    final_article = final_article.strip()
    
    # putting it together
    cikkek_adatai.append([date, title, final_article, m])
    end = time.time()
    counter += 1
    print(str(counter) + '/' + str(len(links)), end - start)

```

```python
df = pd.DataFrame(data=cikkek_adatai)
df.columns = ['dátum','cím', 'szöveg', 'link']
with pd.ExcelWriter('spar_glo.xlsx') as writer:  
    df.to_excel(writer, sheet_name='glo')
```
