```python
import requests
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
import time
```

```python
link_list = []
for n in range(1, 18):
    link = 'https://www.tescoplc.com/media/press-releases/?year=0&categoryId=0&author=0&page=' + str(n)
    r1 = requests.get(link)
    coverpage = r1.content
    soup1 = BeautifulSoup(coverpage, 'html5lib')
    coverpage_news = soup1.find_all('li', class_="listing__item")
    for l in range(len(coverpage_news)):
        link_list.append("https://www.tescoplc.com/" +  coverpage_news[l].find('a')['href'])

link_list
```

### Van egy listám, ami minden idevágó cikk linkjét tartalmazza. (link_list)

```python
cikkek_adatai = []
start = time.time()
counter = 0
for m in link_list:
    # link
    article = requests.get(m)
    article_content = article.content
    soup_article = BeautifulSoup(article_content, 'html5lib')

    # dátum
    date_raw = soup_article.select(".article-date")
    date = " "
    if len(date_raw)>0: 
        date = date_raw[0].get_text()

    # cím
    title_raw = soup_article.select("h1")
    title = title_raw[0].get_text()

    # törzsszöveg
    body = soup_article.select(".rte-wrapper")
    x = body[0].select('p')
    list_paragraphs = []
    for p in np.arange(1, len(x)):
        paragraph = x[p].get_text()
        list_paragraphs.append(paragraph)
        final_article = " ".join(list_paragraphs)
    
    # putting it together
    cikkek_adatai.append([date, title, final_article, m])
    end = time.time()
    counter += 1
    print(str(counter)+"/170", end - start)

```

```python
df = pd.DataFrame(data=cikkek_adatai)
df.columns = ['dátaum', 'cím', 'szöveg', 'link']
with pd.ExcelWriter('tesco_glob.xlsx') as writer:  
    df.to_excel(writer, sheet_name='glo')
```
