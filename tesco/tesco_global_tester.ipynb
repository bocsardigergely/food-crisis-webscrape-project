```python
import requests
from bs4 import BeautifulSoup
import numpy as np
```

```python
r1 = requests.get('https://www.tescoplc.com/media/press-releases/?year=0&categoryId=0&author=0&page=1')
coverpage = r1.content
soup1 = BeautifulSoup(coverpage, 'html5lib')
soup1
```

```python
coverpage_news = soup1.find_all('li', class_="listing__item")
coverpage_news[1]
```

```python
link = "https://www.tescoplc.com/" +  coverpage_news[1].find('a')['href']
article = requests.get(link)
```

```python
article_content = article.content
soup_article = BeautifulSoup(article_content, 'html5lib')
soup_article
```

```python
date_raw = soup_article.select(".article-date")
date = date_raw[0].get_text()
date
```

```python
body = soup_article.select(".rte-wrapper")
body[0]
x = body[0].select('p')
```

```python
list_paragraphs = []
for p in np.arange(1, len(x)):
    paragraph = x[p].get_text()
    list_paragraphs.append(paragraph)
    final_article = " ".join(list_paragraphs)

final_article
```
